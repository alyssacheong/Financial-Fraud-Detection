```python
#This is our code which we used to predict whether a transaction is fraudulent or not
#This first block of code is primarily importing libraries and ensuring our data can be imported easily in the future.

import numpy as np
import pandas as pd
from sklearn import linear_model

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
# Importing the training data into the code so that we can work with it
training_data_file_path = '/kaggle/input/trainingdata8/Data Devils Training Data - Sheet1 (4).csv'
training_data = pd.read_csv(training_data_file_path)

# Trimming down the data into only the columns that we need.
features = ['IsWithdrawal','HoursSince6am','IsMidnight']
X = training_data[features]
y = training_data.IsFraud

#Training and fitting the logistic regression model with training data
logr=linear_model.LogisticRegression()
logr.fit(X,y)

#Importing the test data and trimming down the data to only what we need
real_data_file_path= '/kaggle/input/testdata2/Data Devils Training Data - TestData (1).csv'
real_data = pd.read_csv(real_data_file_path)
real_data_X=real_data[features]

#Running the model on the real test data and storing the predictions
predicted=logr.predict(real_data_X)

#Putting our data into a table and exporting it into a csv file which we can submit
final_table = pd.DataFrame({'TransactionNumber':real_data['TransactionNumber'],
                       'IsFraud':predicted,})
final_table.to_csv('final_table.csv')
```
